```python
from textblob import TextBlob
import pandas as pd
```

```python
cleaned_news = pd.read_csv('data/cleaned_news.csv')
cleaned_news = cleaned_news[['sentences','cleaned_text','date']]
cleaned_news
```

## TextBlob 
TextBlob is an open-source python library for processing textual data which offers a simple API to access its methods and perform basic NLP tasks. TextBlob performs different operations on textual data such as noun phrase extraction, sentiment analysis, classification and translation. Running TextBlob will show subjectivity and polarity of each tweet 

Subjectivity refers to the degree to which a person is personally involved in an object. What matters the most here are personal connections and individual experiences with that object, which may or may not differ from someone else’s point of view. Polarity refers to the strength of an opinion. It could be positive or negative. 

```python
def getSubjectivity(txt):
    return TextBlob(txt).sentiment.subjectivity

def getPolarity (txt):
    return TextBlob(txt).sentiment.polarity
```

```python
cleaned_news['Subjectivity']= cleaned_news['sentences'].apply(getSubjectivity)
cleaned_news['Polarity'] = cleaned_news['sentences'].apply(getPolarity)
cleaned_news.head()
```

```python
# Filter rows where Polarity is less than 0
negative_polarity_df = cleaned_news[cleaned_news['Polarity'] < 0]

# Display the filtered DataFrame
negative_polarity_df
```

## Transformers

```python
import pandas as pd
from transformers import pipeline

# Initialize the sentiment-analysis pipeline
sentiment_pipeline = pipeline("sentiment-analysis")

# Applying the sentiment-analysis model
cleaned_news['sentiment'] = cleaned_news['sentences'].apply(lambda text: sentiment_pipeline(text)[0])

# View results
cleaned_news
```

## Vader

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download the VADER lexicon
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()
```

```python
# Apply VADER to each text and store the compound score
cleaned_news['vader_score'] = cleaned_news['sentences'].apply(lambda text: sia.polarity_scores(text)['compound'])

cleaned_news
```

```python
from joblib import dump, load
loaded_model = load('models/voting_classifier_model.joblib')
print("Model loaded")
loaded_vectorizer = load('models/tfidf_vectorizer.joblib')
print("Vecotrizer loaded")

# Transform the text data using the loaded vectorizer
transformed_texts = loaded_vectorizer.transform(cleaned_news['cleaned_text'])
```

```python
# Get predictions
predictions = loaded_model.predict(transformed_texts)
print("Predictions:", predictions)
```

```python
# Add predictions to the DataFrame
cleaned_news['predicted_class'] = predictions
cleaned_news
```

```python
cleaned_news = cleaned_news[cleaned_news['predicted_class'] != 'General']
cleaned_news
```

```python
cleaned_news['date'] = pd.to_datetime(cleaned_news['date'])
# To calculate the average daily sentiment score
daily_category_avg_score = cleaned_news.groupby(['date', 'predicted_class'])['vader_score'].mean().reset_index()
print("Average Daily Sentiment:")
daily_category_avg_score
```

```python
daily_category_avg_score.to_csv('data/ESG_news_scores.csv')
```
