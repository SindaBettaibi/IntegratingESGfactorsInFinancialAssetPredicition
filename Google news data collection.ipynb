```python
import datetime
import requests
import pandas as pd
from gnews import GNews
from tqdm.auto import tqdm

from importlib import reload
import text_cleaner  # Import the module first
text_cleaner = reload(text_cleaner)  # Reload the module
from text_cleaner import TextCleaner  # Re-import the class
from sentence_extractor import SentenceExtractor
```

## Scraping google news articles

```python
keywords = ["Microsoft sustainability", "Microsoft ESG"]
start_year = 2018
end_year = 2023

# List to hold each search's DataFrame
dataframes = []
for keyword in keywords:
    for year in range(start_year, end_year + 1):
        # Define the start and end dates for each year
        start_date = datetime.date(year, 1, 1)
        end_date = datetime.date(year, 12, 31)

        # Initialize GNews object for the current year with specified dates
        google_news = GNews(language='en', start_date=start_date, end_date=end_date)

        # Fetch news articles for the year based on the current keyword
        yearly_news = google_news.get_news(keyword)

        # Check if news articles exist for the year and append to list
        if yearly_news:
            df = pd.DataFrame(yearly_news)
            dataframes.append(df)

# Concatenate all DataFrames into one DataFrame
full_df = pd.concat(dataframes, ignore_index=True)
full_df
```

```python
df1 =full_df.copy()
```

```python
# Remove 'description' and 'url' columns
df1.drop(['description'], axis=1, inplace=True)

# Convert 'published date' to only show the year
df1['published date'] = pd.to_datetime(df1['published date'], errors='coerce', format='%a, %d %b %Y %H:%M:%S GMT')
df1['published date'] = df1['published date'].dt.strftime('%d-%m-%Y')  # This will create a new column with the formatted date

# Extract the 'title' from the 'publisher' dictionary and create a new column 'source'
df1['source'] = df1['publisher'].apply(lambda x: x['title'] if isinstance(x, dict) and 'title' in x else None)
df1.rename(columns={'published date': 'date'}, inplace=True)

# Remove the original 'publisher' column
df1.drop('publisher', axis=1, inplace=True)
df1
```

## Articles URLs content extraction

```python
def fetch_and_extract(client, url):
    try:
        article = client.get_full_article(url)
        if article and hasattr(article, 'text'):
            return article.text
        else:
            return None  # or return "" if you prefer to keep type consistency
    except Exception as e:
        print(f"Failed to fetch content from {url}: {e}")
        return None

google_news = GNews()
merged_df['content'] = merged_df['url'].apply(lambda x: fetch_and_extract(google_news, x))
merged_df['content'] = merged_df['content'].fillna(merged_df['title'])
merged_df
```

## Sentence Extraction

```python
processor = SentenceExtractor()
# Apply the extract_sentences function to the 'content' column
merged_df['sentences'] = merged_df['content'].apply(lambda x: processor.extract_sentences(x))

# Explode the 'sentences' list into separate rows
df_exploded = merged_df.explode('sentences')

# Create a new DataFrame focusing on the normalized data
result_df = df_exploded[['title', 'date', 'source', 'sentences']]
result_df['sentences'] = result_df['sentences'].fillna(result_df['title'])

result_df.reset_index(drop=True, inplace=True)

result_df
```

```python
result_df['sentences'] = result_df['sentences'].fillna(result_df['title'])
```

## Text cleaning

```python
cleaner = TextCleaner() 
# Apply the text-cleaning methods
result_df['sentences'] = result_df['sentences'].apply(cleaner.remove_special_characters)
result_df['cleaned_text'] = result_df['sentences'].apply(cleaner.lowercase_text)
result_df['cleaned_text'] = result_df['cleaned_text'].apply(cleaner.remove_entities)
result_df['cleaned_text'] = result_df['cleaned_text'].apply(cleaner.remove_punctuation)
result_df['cleaned_text'] = result_df['cleaned_text'].apply(cleaner.remove_stopwords)
result_df
```

```python
cleaned_news = result_df[['sentences','cleaned_text','date','source']]
cleaned_news.to_csv('data/cleaned_news.csv')
```
